# Setup

# Inferencing
This cell demonstrated how we can use the model object as well as the created access token
to pair it with parameters and input string to obtain
the response from the the selected foundation model.

## Defining the model id
We need to specify model id that will be used for inferencing:


```python
model_id = "ibm/granite-13b-chat-v2"
```

## Defining the model parameters
We need to provide a set of model parameters that will influence the
result:

```python
parameters = {
    "decoding_method": "greedy",
    "max_new_tokens": 200,
    "repetition_penalty": 1
}
```

## Defining the project id or space id
The API requires project id or space id that provides the context for the call. We will obtain
the id from the project or space in which this notebook runs:

```python
project_id = os.getenv("PROJECT_ID")
space_id = os.getenv("SPACE_ID")
```

## Defining the Model object
We need to define the Model object using the properties we defined so far:


```python
from ibm_watsonx_ai.foundation_models import Model

model = Model(
	model_id = model_id,
	params = parameters,
	credentials = get_credentials(),
	project_id = project_id,
	space_id = space_id
	)
```

## Defining the inferencing input
Foundation model inferencing API accepts a natural language input that it will use
to provide the natural language response. The API is sensitive to formatting. Input
structure, presence of training steps (one-shot, two-shot learning etc.), as well
as phrasing all influence the final response and belongs to the emerging discipline of
Prompt Engineering.

Let us provide the input we got from the Prompt Lab:


```python
prompt_input = """summarize the problem in the input(feedback)  and help with top one solution that can be implemented to solve the problem.

Input: 1.The instructor seemed inexperienced and struggled to provide clear answers to our questions. Their lack of expertise was evident and hindered the learning process. I felt the instructor didn’t add much value to the session.

2.The training sessions were monotonous and lacked engagement. There were no interactive elements to keep us involved, making it difficult to stay focused. It felt like we were just being talked at for hours.

3.The presentation was rushed and hard to follow. The instructor moved too quickly through important points without ensuring we understood. This left many of us confused and unable to keep up.
Output: Provide additional training for instructors to enhance their subject matter expertise and teaching skills. Incorporate interactive elements such as Q&A sessions and hands-on activities to keep participants engaged and ensure a manageable pace with regular checks for understanding.

Input: 1. The instructor's explanations were too technical and difficult to understand. They didn't simplify the concepts for beginners.

2. There was no opportunity for us to apply what we learned through practical exercises. The training was too theoretical.

3. The sessions were too long without any breaks, making it hard to maintain concentration and absorb the information.

Output: Introduce a more balanced training approach by simplifying technical explanations, incorporating practical exercises to apply learned concepts, and scheduling regular breaks to help maintain focus and improve information retention.

Input: 1. Our team meetings often lack diverse perspectives, as some employees are hesitant to share their ideas due to fear of judgment.
2. There is a noticeable lack of representation in leadership positions, which makes it hard for underrepresented groups to feel motivated and included.
3. Despite having diversity policies, there is little follow-through, and many employees feel that their concerns about inclusion are not being taken seriously.


Output: Implement a comprehensive inclusion initiative that encourages open dialogue by creating safe spaces for sharing ideas, establishes mentorship programs to support diverse talent in leadership roles, and ensures accountability by regularly reviewing and acting on feedback related to inclusion concerns.

Input: 1. The expectation to be available outside of regular working hours is affecting my personal time and causing burnout.

2. There is no clear boundary between work and personal life, especially with remote work. It feels like I'm always on the clock.

3. The workload is overwhelming, and there is little support or flexibility from management to manage personal responsibilities.

Output:"""
```

## Execution
Let us now use the defined Model object and pair it with input and
generate the response:


```python
print("Submitting generation request...")
generated_response = model.generate_text(prompt=prompt_input, guardrails=True)
print(generated_response)
```
